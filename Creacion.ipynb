# Cassava, fresh database

```
import pandas as pd
```

## Crops and livestock productions

```
crop_liv_prodc = pd.read_csv('./Crops and livestock products/Crops and livestock products.csv')
# Only useful columns
crop_liv_prodc.drop(['Domain Code', 'Domain', 'Element Code',
    'Item Code (CPC)', 'Item', 'Year Code', 'Unit',
       'Flag', 'Flag Description'], axis=1, inplace=True)
crop_pivot = crop_liv_prodc.pivot(index=['Area', 'Area Code (M49)'], columns='Element', values='Value').reset_index()
#Change the names to see the unit
crop_pivot = crop_pivot.rename(columns={'Area harvested': 'Area harvested (ha)'})
crop_pivot = crop_pivot.rename(columns={'Yield': 'Yield (hg/ha)'})
crop_pivot = crop_pivot.rename(columns={'Production': 'Production (tonnes)'})
```

```
#Final
crop_pivot
```

## Producer prices

```
#Original
prod_prices = pd.read_csv('./Producer prices/Producer prices.csv')
#Innecesary columns
prod_prices.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Item Code (CPC)', 'Item', 'Year Code',
       'Months Code', 'Months', 'Unit', 'Flag', 'Flag Description'], axis=1, inplace=True)
#One per country
prod_prices_pivot = prod_prices.pivot(index='Area', columns='Element', values='Value').reset_index()
```

```
#Final
prod_prices_pivot
```

## Production indices

```
#Original
indx_prod = pd.read_excel('./Production indices/Production indices.xlsx')
#Innecesary columns
indx_prod.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Item Code (CPC)', 'Item', 'Year Code', 'Unit',
    'Flag', 'Flag Description'], axis=1, inplace=True)
#One per country
indx_prod_pivot = indx_prod.pivot(index='Area', columns='Element', values='Value').reset_index()
```

```
#Final
indx_prod_pivot
```

## Trade. Crops and livestock

```
#Original
trad_cro_liv = pd.read_csv('./Trade. Crops and livestock/Trade. Crops and livestock.csv')
#Innecesary columns
trad_cro_liv.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Item Code (CPC)', 'Item', 'Year Code', 'Unit',
       'Flag', 'Flag Description'], axis=1, inplace=True)
#One per country
trad_cro_liv_pivot = trad_cro_liv.pivot(index='Area', columns='Element', values='Value').reset_index()
```

```
#Final
trad_cro_liv_pivot
```

## Value of agriculture production

```
#Original
val_ag_prod = pd.read_csv('./Value of agricultural production/Value of agricultural production.csv')
#Innecesary columns
val_ag_prod.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Item Code (CPC)', 'Item', 'Year Code', 'Unit',
       'Flag', 'Flag Description'], axis=1, inplace=True)
#One per country
val_ag_prod_pivot = val_ag_prod.pivot(index='Area', columns='Element', values='Value').reset_index()
```

```
#Final
val_ag_prod_pivot.describe()
```

## Population

```
#Original
population = pd.read_csv('./Population/Annual population.csv')
#Innecesary columns
population.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Item', 'Item Code', 'Year Code', 'Unit',
       'Flag', 'Flag Description', 'Note'], axis=1, inplace=True)
#One per country
populatio_pivot = population.pivot(index='Area', columns='Element', values='Value').reset_index()
```

```
#Final
populatio_pivot
```

## Land Use

```
#Original
land_use = pd.read_csv('./Land Use/Land Use.csv')
#Innecesary columns 
land_use.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Element', 'Item Code', 'Year Code', 'Year', 'Unit',
       'Flag', 'Flag Description'], axis=1, inplace=True)
#One per country
land_use_pivot = land_use.pivot(index='Area', columns='Item', values='Value').reset_index()
```

```
land_use_pivot
```

## Gov expenditure

```
#Original
gov_expenditure = pd.read_csv('./Government Expenditure/gov_expenditure1.csv')
#Innecesary columns: quedan: area, 
gov_expenditure.drop(['Domain Code', 'Domain', 'Area Code (M49)', 'Element Code',
       'Element', 'Item Code', 'Year Code', 'Unit', 'Year',
       'Flag', 'Flag Description', 'Note'], axis=1, inplace=True)
#One per country
gov_expenditure_pivot = gov_expenditure.pivot(index='Area', columns='Item', values='Value').reset_index()
```

## The big df

```
#Juntar cassava
total = pd.merge(crop_pivot, prod_prices_pivot, on='Area', how='outer')
total = pd.merge(total, indx_prod_pivot, on='Area', how='outer')
total = pd.merge(total, trad_cro_liv_pivot, on='Area', how='outer')
total = pd.merge(total, val_ag_prod_pivot, on='Area', how='outer')
#Generales
total = pd.merge(total, populatio_pivot, on='Area', how='inner')
total = pd.merge(total, land_use_pivot, on='Area', how='inner')
total = pd.merge(total, gov_expenditure_pivot, on='Area', how='left')
```

```
total.describe()
```

## Additional categories

```
country_gropus = pd.read_excel('/work/Geral/Continents.xlsx')
#Additional categories
total = pd.merge(total, country_gropus[['Country', 'Country Group']], left_on='Area', right_on='Country', how='left')
#Delete
total.drop('Country', axis=1, inplace=True)
```
